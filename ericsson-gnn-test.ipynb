```python
class GraphFromDataFrame:
    def __init__(self, nodes_df, edges_df):
        self.nodes = nodes_df['func_id'].tolist()
        self.node_index = {node: i for i, node in enumerate(self.nodes)}  # mapping node id -> index
        self.adj = defaultdict(list)
        self.V = len(self.nodes)
        
        # Élek hozzáadása
        for _, row in edges_df.iterrows():
            src = row['source_id']
            tgt = row['target_id']
            self.add_edge(src, tgt)

    def add_edge(self, v, w):
        self.adj[v].append(w)
        self.adj[w].append(v)  # mert irányítatlan gráf

    def dfs(self, v, visited):
        visited.add(v)
        for neighbor in self.adj[v]:
            if neighbor not in visited:
                self.dfs(neighbor, visited)

    def number_of_connected_components(self):
        visited = set()
        count = 0
        for node in self.nodes:
            if node not in visited:
                self.dfs(node, visited)
                count += 1
        return count



g = GraphFromDataFrame(cg_nodes, cg_edges)
print(g.number_of_connected_components())
```

# Autoencoder

```python
class HeteroGraphAutoEncoder(nn.Module):
    def __init__(self):
        super().__init__()

        # Dimenziók: function: 384 → 256 → 128, expression: 768 → 512 → 256 → 128
        self.encoder_layers = nn.ModuleList()

        # 1. Function → Function
        self.encoder_layers.append(HeteroConv({
            ('function', 'calls', 'function'): SAGEConv((384, 384), 256)
        }, aggr='sum'))

        # 2. Expression → Expression
        self.encoder_layers.append(HeteroConv({
            ('expression', 'connected_to', 'expression'): SAGEConv((768, 768), 512)
        }, aggr='sum'))

        self.encoder_layers.append(HeteroConv({
            ('expression', 'connected_to', 'expression'): SAGEConv((512, 512), 256)
        }, aggr='sum'))

        # 3. Function ↔ Expression (contains & is_in)
        self.encoder_layers.append(HeteroConv({
            ('function', 'contains', 'expression'): SAGEConv((256, 256), 128),
            ('expression', 'is_in', 'function'): SAGEConv((256, 256), 128),
        }, aggr='sum'))

        # 4. Function → Function
        self.encoder_layers.append(HeteroConv({
            ('function', 'calls', 'function'): SAGEConv((128, 128), 128)
        }, aggr='sum'))

        # 5. Expression → Expression
        self.encoder_layers.append(HeteroConv({
            ('expression', 'connected_to', 'expression'): SAGEConv((128, 128), 128)
        }, aggr='sum'))

        self.encoder_layers.append(HeteroConv({
            ('expression', 'connected_to', 'expression'): SAGEConv((128, 128), 128)
        }, aggr='sum'))

        # Decoder (rekonstruálja az eredeti node feature-öket)
        self.decoder_func = nn.Sequential(
            Linear(128, 256),
            nn.ReLU(),
            Linear(256, 384)
        )
        self.decoder_expr = nn.Sequential(
            Linear(128, 512),
            nn.ReLU(),
            Linear(512, 768)
        )

    def encode(self, data: HeteroData):
        x_dict, edge_index_dict = data.x_dict, data.edge_index_dict

        for i, conv in enumerate(self.encoder_layers):
            if conv.convs.keys() == [('function', 'calls', 'function')]:
                x_dict['function'] = F.relu(conv(x_dict, edge_index_dict)['function'])
            elif conv.convs.keys() == [('expression', 'connected_to', 'expression')]:
                x_dict['expression'] = F.relu(conv(x_dict, edge_index_dict)['expression'])
            elif conv.convs.keys() == [('function', 'contains', 'expression'), ('expression', 'is_in', 'function')]:
                x_dict = conv(x_dict, edge_index_dict)
                x_dict['function'] = F.relu(x_dict['function'])
                x_dict['expression'] = F.relu(x_dict['expression'])
        return x_dict

    def decode(self, z_dict):
        out = {
            'function': self.decoder_func(z_dict['function']),
            'expression': self.decoder_expr(z_dict['expression']),
        }
        return out

    def forward(self, data: HeteroData):
        z_dict = self.encode(data)
        recon_dict = self.decode(z_dict)
        return recon_dict, z_dict
```

```python
# Model 
model = HeteroGraphAutoEncoder().to('cuda' if torch.cuda.is_available() else 'cpu')
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)

# Data to GPU
device = next(model.parameters()).device
data = data.to(device)

# Save original features
x_orig_dict = {
    'function': data['function'].x.clone().detach(),
    'expression': data['expression'].x.clone().detach(),
}

# Learning
for epoch in range(1, 201):
    model.train()
    optimizer.zero_grad()

    x_recon_dict, z_dict = model(data)

    # MSE Loss
    loss_MSE = (
        F.mse_loss(x_recon_dict['function'], x_orig_dict['function']) +
        F.mse_loss(x_recon_dict['expression'], x_orig_dict['expression'])
    )

    # Cosine Similarity Loss
    sim_func = F.cosine_similarity(x_recon_dict['function'], x_orig_dict['function'], dim=1)
    sim_expr = F.cosine_similarity(x_recon_dict['expression'], x_orig_dict['expression'], dim=1)

    cosine_score_func = sim_func.mean()
    cosine_score_expr = sim_expr.mean()

    loss_cosine = (cosine_score_func + cosine_score_expr) / 2

    loss_MSE.backward()
    optimizer.step()

    print(f"[{epoch:03d}] MSE: {loss_MSE.item():.4f}, Cosine Similarity: {loss_cosine.item():.4f}")
```

```python
# GPU ellenőrzés
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Modell betöltése (csak encoder kell)
tokenizer = BartTokenizer.from_pretrained("facebook/bart-base")
BART = BartModel.from_pretrained("facebook/bart-base").to(device)
BART.eval()

MAX_TOKENS = 64

# Lista az új oszlopokhoz
all_full_embeddings = []

token_lengths = []

# Kódolás minden sorra
for docstring in tqdm(cg_nodes['docstring']):
    if pd.isna(docstring) or not isinstance(docstring, str) or len(docstring.strip()) == 0:
        # Ha a docstring NaN vagy üres, akkor nullvektorral helyettesítjük
        full_embedding = np.zeros((MAX_TOKENS, 768)).tolist()  # [256, 768] nullvektor
        token_lengths.append(0)
        all_full_embeddings.append(full_embedding)
        continue

    inputs = tokenizer(docstring, return_tensors="pt", truncation=True, max_length=MAX_TOKENS).to(device)
    
    with torch.no_grad():
        encoder_outputs = BART.encoder(**inputs)  # BaseModelOutput
        hidden_states = encoder_outputs.last_hidden_state  # [1, seq_len, 768]

        seq_len = hidden_states.size(1)
        if seq_len < MAX_TOKENS:
            # Paddoljuk nullvektorokkal a [1, seq_len, 768] -> [1, 256, 768]
            padding = torch.zeros((1, MAX_TOKENS - seq_len, hidden_states.size(2)), device=hidden_states.device)
            hidden_states = torch.cat([hidden_states, padding], dim=1)


        full_embedding = hidden_states.squeeze(0).cpu().tolist()  # [seq_len, 768]

        token_lengths.append(len(full_embedding))

    all_full_embeddings.append(full_embedding)

# Új oszlopok hozzáadása
cg_nodes['docstring_BART_embedding'] = all_full_embeddings

# Mentés fájlba (ha kell)
cg_nodes
```

```python
class EmbeddingAutoencoder(nn.Module):
    def __init__(self):
        super(EmbeddingAutoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(49152, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.Tanh(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 2048),
            nn.LeakyReLU(),
            nn.Linear(2048, 768)
        )
        self.decoder = nn.Sequential(
            nn.Linear(768, 2048),
            nn.LeakyReLU(),
            nn.Linear(2048, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.Tanh(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 49152),
        )

    def forward(self, x):
        z = self.encoder(x)
        x_hat = self.decoder(z)
        return x_hat

    def encode(self, x):
        return self.encoder(x)
```

```python
class AutoencoderDataset(Dataset):
    def __init__(self, df):
        # A 64x768-as embeddinget flatteneljük: 64 * 768 = 49152
        flattened = [np.array(x).flatten() for x in df['docstring_BART_embedding'].values]
        self.data = torch.tensor(np.vstack(flattened), dtype=torch.float32)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        x = self.data[idx]
        return x
```

### Docstring pred GNN Mask nodes

```python
def create_function_train_test_masks(data: HeteroData, train_ratio=0.8, seed=42):
    """
    Csak azokat a function node-okat maszkoljuk, amelyekhez tartozik docstring.
    Ezeket szétosztjuk train és test maszkolásra.
    """
    func_x = data['function'].x
    is_not_empty = func_x.norm(p=2, dim=1) > 0
    valid_indices = is_not_empty.nonzero(as_tuple=False).squeeze()

    torch.manual_seed(seed)
    perm = torch.randperm(valid_indices.size(0))
    num_train = int(train_ratio * valid_indices.size(0))

    train_indices = valid_indices[perm[:num_train]]
    test_indices = valid_indices[perm[num_train:]]

    train_mask = torch.zeros(data['function'].num_nodes, dtype=torch.bool)
    test_mask = torch.zeros(data['function'].num_nodes, dtype=torch.bool)

    train_mask[train_indices] = True
    test_mask[test_indices] = True

    return train_mask, test_mask




train_mask, test_mask = create_function_train_test_masks(data, train_ratio=0.8, seed=42)

original_function_x = data['function'].x.clone()

masked_function_x = original_function_x.clone()
masked_function_x[train_mask] = 0.0  # csak train maszkolva
data['function'].x = masked_function_x
```

```python
class DocstringPredictorGNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.convs = nn.ModuleList()

        self.convs.append(HeteroConv({
            ('function', 'calls', 'function'): SAGEConv((768, 768), 256)
        }, aggr='sum'))

        self.convs.append(HeteroConv({
            ('expression', 'connected_to', 'expression'): SAGEConv((768, 768), 512)
        }, aggr='sum'))

        self.convs.append(HeteroConv({
            ('expression', 'connected_to', 'expression'): SAGEConv((512, 512), 256)
        }, aggr='sum'))

        self.convs.append(HeteroConv({
            ('function', 'contains', 'expression'): SAGEConv((256, 256), 128),
            ('expression', 'is_in', 'function'): SAGEConv((256, 256), 128)
        }, aggr='sum'))

        self.convs.append(HeteroConv({
            ('function', 'calls', 'function'): SAGEConv((128, 128), 128)
        }, aggr='sum'))

        self.convs.append(HeteroConv({
            ('expression', 'connected_to', 'expression'): SAGEConv((128, 128), 128)
        }, aggr='sum'))

        self.convs.append(HeteroConv({
            ('expression', 'connected_to', 'expression'): SAGEConv((128, 128), 128)
        }, aggr='sum'))

        self.convs.append(HeteroConv({
            ('expression', 'is_in', 'function'): SAGEConv((128, 128), 128)
        }, aggr='sum'))

        # Final MLP to predict docstring embedding (384 dim)
        self.predictor = nn.Sequential(
            nn.Linear(128, 256),
            nn.LeakyReLU(),
            nn.Linear(256, 384),
            nn.LeakyReLU(),
            nn.Linear(384, 768)
        )



    def forward(self, data: HeteroData):
        x_dict, edge_index_dict = data.x_dict, data.edge_index_dict

        for i, conv in enumerate(self.convs):
            if conv.convs.keys() == [('function', 'calls', 'function')]:
                x_dict['function'] = F.leaky_relu(conv(x_dict, edge_index_dict)['function'])
            elif conv.convs.keys() == [('expression', 'connected_to', 'expression')]:
                x_dict['expression'] = F.leaky_relu(conv(x_dict, edge_index_dict)['expression'])
            elif conv.convs.keys() == [('function', 'contains', 'expression'), ('expression', 'is_in', 'function')]:
                x_dict = conv(x_dict, edge_index_dict)
                x_dict['function'] = F.leaky_relu(x_dict['function'])
                x_dict['expression'] = F.leaky_relu(x_dict['expression'])
            elif conv.convs.keys() == [('expression', 'isin', 'function')]:
                x_dict['function'] = F.leaky_relu(conv(x_dict, edge_index_dict)['function'])


        out = self.predictor(x_dict['function'])
        return out
```

```python
# Data to GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
data = data.to(device)

model = DocstringPredictorGNN().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-2)


for epoch in range(1, 51):
    model.train()
    optimizer.zero_grad()

    pred = model(data)

    target_train = original_function_x[train_mask]
    pred_train = pred[train_mask]

    loss_cos = 1 - cosine_similarity(pred_train.to(device), target_train.to(device), dim=1).mean()
    loss_MSE = F.mse_loss(pred_train.to(device), target_train.to(device))

    loss_MSE.backward()
    optimizer.step()

    # Teszt kiértékelés (nem használódik fel tanításra)
    model.eval()
    with torch.no_grad():
        target_test = original_function_x[test_mask]
        pred_test = pred[test_mask]

        test_cos = 1 - cosine_similarity(pred_test.to(device), target_test.to(device), dim=1).mean()
        test_MSE = F.mse_loss(pred_test.to(device), target_test.to(device))

    print(f"[{epoch:03d}] Train Cos: {loss_cos.item():.4f}, MSE: {loss_MSE.item():.4f} | "
          f"Test Cos: {test_cos.item():.4f}, MSE: {test_MSE.item():.4f}")
```

```python
def docstring_filter(dataset):
    dataset['docstring'] = dataset['docstring'].astype(str)
    dataset = dataset[
        (~dataset['docstring'].isna()) & 
        (dataset['docstring'] != "") & 
        (dataset['docstring'] != "None") &
        (dataset['docstring'] != "nan")
    ]
    return dataset.reset_index(drop=True)

sk_cg_nodes = pd.read_csv("./graph/sklearn/cg_nodes.csv")
sk_cg_nodes = docstring_filter(sk_cg_nodes)

mn_cg_nodes = pd.read_csv("./graph/manim/cg_nodes.csv")
mn_cg_nodes = docstring_filter(mn_cg_nodes)

cg_nodes = pd.concat([sk_cg_nodes, mn_cg_nodes], ignore_index=True)[['docstring', 'docstring_embedding']]
cg_nodes
```

```python
# GPU ellenőrzés
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Modell betöltése (csak encoder kell)
tokenizer = BartTokenizer.from_pretrained("facebook/bart-base")
model = BartModel.from_pretrained("facebook/bart-base").to(device)
model.eval()

MAX_TOKENS = 64

# Lista az új oszlopokhoz
all_full_embeddings = []
all_mean_embeddings = []

token_lengths = []

# Kódolás minden sorra
for docstring in tqdm(cg_nodes['docstring']):
    if pd.isna(docstring) or not isinstance(docstring, str) or len(docstring.strip()) == 0:
        print("Skipping empty or invalid docstring")
        continue

    inputs = tokenizer(docstring, return_tensors="pt", truncation=True, max_length=MAX_TOKENS).to(device)
    
    with torch.no_grad():
        encoder_outputs = model.encoder(**inputs)  # BaseModelOutput
        hidden_states = encoder_outputs.last_hidden_state  # [1, seq_len, 768]

        seq_len = hidden_states.size(1)
        if seq_len < MAX_TOKENS:
            # Paddoljuk nullvektorokkal a [1, seq_len, 768] -> [1, 256, 768]
            padding = torch.zeros((1, MAX_TOKENS - seq_len, hidden_states.size(2)), device=hidden_states.device)
            hidden_states = torch.cat([hidden_states, padding], dim=1)


        mean_embedding = hidden_states.mean(dim=1).squeeze(0).cpu().tolist()  # [768]
        full_embedding = hidden_states.squeeze(0).cpu().tolist()  # [seq_len, 768]

        token_lengths.append(len(full_embedding))

    all_full_embeddings.append(full_embedding)
    all_mean_embeddings.append(mean_embedding)

# Új oszlopok hozzáadása
cg_nodes['docstring_embedding'] = all_full_embeddings
cg_nodes['docstring_mean_embedding'] = all_mean_embeddings

# Mentés fájlba (ha kell)
cg_nodes
```

```python
class AutoencoderDataset(Dataset):
    def __init__(self, df):
        # A 64x768-as embeddinget flatteneljük: 64 * 768 = 49152
        flattened = [np.array(x).flatten() for x in df['docstring_embedding'].values]
        self.data = torch.tensor(np.vstack(flattened), dtype=torch.float32)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        x = self.data[idx]
        return x
```

```python
class EmbeddingAutoencoder(nn.Module):
    def __init__(self):
        super(EmbeddingAutoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(49152, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 2048),
            nn.LeakyReLU(),
            nn.Linear(2048, 768)
        )
        self.decoder = nn.Sequential(
            nn.Linear(768, 2048),
            nn.LeakyReLU(),
            nn.Linear(2048, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 49152),
        )

    def forward(self, x):
        z = self.encoder(x)
        x_hat = self.decoder(z)
        return x_hat
```

```python
# Feltételezzük, hogy df a betöltött DataFrame-ed, amely tartalmazza a két szükséges oszlopot
dataset = AutoencoderDataset(cg_nodes)
```

```python
# GPU beállítás
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Train-test split (80% train, 20% test)
dataset_size = len(dataset)
train_size = int(0.8 * dataset_size)
test_size = dataset_size - train_size

train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

batch_size = 192

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Modell példányosítás és GPU-ra küldés
model = EmbeddingAutoencoder().to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=8.5e-5)

criterion_mse = torch.nn.MSELoss()
cosine_sim = torch.nn.CosineSimilarity(dim=1)

for epoch in range(10):

    model.train()
    running_cosine_loss = 0.0
    running_mse_loss = 0.0
    running_loss = 0.0

    for data in train_loader:
        
        data = data.to(device)
        optimizer.zero_grad()
        outputs = model(data)
        
        cosine_loss = 1 - cosine_sim(outputs, data).mean()
        mse_loss = criterion_mse(outputs, data)

        loss = 1.0 * cosine_loss + 0.6 * mse_loss

        loss.backward()
        optimizer.step()

        running_cosine_loss += cosine_loss.item() * data.size(0)
        running_mse_loss += mse_loss.item() * data.size(0)
        running_loss += loss.item() * data.size(0)


    epoch_train_loss = running_loss / train_size
    epoch_train_cosine_loss = running_cosine_loss / train_size
    epoch_train_mse_loss = running_mse_loss / train_size

    # Teszt fázis
    model.eval()
    running_test_loss = 0.0
    running_test_cosine_loss = 0.0
    running_test_mse_loss = 0.0

    with torch.no_grad():
        for data in test_loader:
            data = data.to(device)
            outputs = model(data)
            cosine_loss = 1 - cosine_sim(outputs, data).mean()
            mse_loss = criterion_mse(outputs, data)
            loss = 1.0 * cosine_loss + 0.6 * mse_loss
            running_test_loss += loss.item() * data.size(0)
            running_test_cosine_loss += cosine_loss.item() * data.size(0)
            running_test_mse_loss += mse_loss.item() * data.size(0)

    epoch_test_loss = running_test_loss / test_size

    print(f"""Epoc {epoch+1}\n
              Train Loss: {epoch_train_loss:.6f} | Train Cosine Loss: {epoch_train_cosine_loss:.6f} | Train MSE Loss: {epoch_train_mse_loss:.6f}\n
              Test Loss: {epoch_test_loss:.6f} | Test Cosine Loss: {running_test_cosine_loss / test_size:.6f} | Test MSE Loss: {running_test_mse_loss / test_size:.6f}\n""")
```

```python
class EmbeddingDataset(Dataset):
    def __init__(self, df):
        self.inputs = torch.tensor(np.vstack(df['docstring_mean_embedding'].values), dtype=torch.float32)

        # Az x most lista, ezért először np.array, majd flatten
        flattened_targets = [np.array(x).flatten() for x in df['docstring_embedding'].values]
        self.targets = torch.tensor(np.vstack(flattened_targets), dtype=torch.float32)

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        return self.inputs[idx], self.targets[idx]
```

```python
class EmbeddingDecoder(nn.Module):
    def __init__(self):
        super(EmbeddingDecoder, self).__init__()
        self.decoder = nn.Sequential(
            nn.Linear(768, 2048),
            nn.LeakyReLU(),
            nn.Linear(2048, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192*2),
            nn.LeakyReLU(),
            nn.Linear(8192*2, 49152),
        )

    def forward(self, x):
        return self.decoder(x)
```

```python
# --- 1) Eszköz kiválasztása --- #
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


# --- 3) Modell és tokenizer --- #
tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small").to(device)
model.eval()

# --- 4) Paraméterek --- #
MAX_LEN = 64
HIDDEN_SIZE = model.config.d_model
BATCH_SIZE = 64

# --- 5) Batch-es embedding függvény --- #
def embed_batch(text_list):
    inputs = tokenizer(
        text_list,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=MAX_LEN,
    ).to(device)  # shape: [batch, seq_len]

    with torch.no_grad():
        encoder_outputs = model.encoder(input_ids=inputs.input_ids, attention_mask=inputs.attention_mask)
        embeddings = encoder_outputs.last_hidden_state  # shape: [batch, seq_len, hidden_size]

    result = []
    for i in range(embeddings.size(0)):
        emb = embeddings[i]  # shape: [seq_len_i, hidden_size]
        seq_len = emb.size(0)

        if seq_len < MAX_LEN:
            pad = torch.zeros(MAX_LEN - seq_len, HIDDEN_SIZE, device=device)
            emb = torch.cat([emb, pad], dim=0)
        else:
            emb = emb[:MAX_LEN, :]

        result.append(emb.cpu())  # CPU-ra vissza
    return result

# --- 6) Alkalmazás batch-ekre bontva --- #
all_embeddings = []
for i in tqdm(range(0, len(df), BATCH_SIZE)):
    batch_texts = df["docstring"].iloc[i:i+BATCH_SIZE].tolist()
    batch_embs = embed_batch(batch_texts)
    all_embeddings.extend(batch_embs)

df["embedding"] = all_embeddings
```

```python
class EmbeddingAutoencoder(nn.Module):
    def __init__(self):
        super(EmbeddingAutoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(32768, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 2048),
            nn.LeakyReLU(),
            nn.Linear(2048, 1024)
        )
        self.decoder = nn.Sequential(
            nn.Linear(1024, 2048),
            nn.LeakyReLU(),
            nn.Linear(2048, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 8192),
            nn.LeakyReLU(),
            nn.Linear(8192, 32768),
        )

    def forward(self, x):
        z = self.encoder(x)
        x_hat = self.decoder(z)
        return x_hat
```

```python
train_loader = DataLoader(train_df.tolist(), batch_size=64, shuffle=True)
test_loader = DataLoader(test_df.tolist(), batch_size=64, shuffle=False)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = EmbeddingAutoencoder().to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3)
mse_loss = nn.MSELoss()
cosine_similarity = nn.CosineSimilarity(dim=1)

num_epochs = 20
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    running_cosine_loss = 0.0
    running_mse_loss = 0.0

    for batch in train_loader:
        batch = batch.to(device)
        optimizer.zero_grad()
        outputs = model(batch)

        loss_cos = 1 - cosine_similarity(outputs, batch).mean()
        loss_mse = mse_loss(outputs, batch)

        loss = 1.0 * loss_cos + 0.6 * loss_mse
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * batch.size(0)
        running_cosine_loss += loss_cos.item() * batch.size(0)
        running_mse_loss += loss_mse.item() * batch.size(0)

    epoch_train_loss = running_loss / len(train_loader.dataset)
    epoch_train_cosine_loss = running_cosine_loss / len(train_loader.dataset)
    epoch_train_mse_loss = running_mse_loss / len(train_loader.dataset)

    # Teszt fázis
    model.eval()
    running_test_loss = 0.0
    running_test_cosine_loss = 0.0
    running_test_mse_loss = 0.0

    with torch.no_grad():
        for batch in test_loader:
            batch = batch.to(device)
            outputs = model(batch)
            loss_cos = 1 - cosine_similarity(outputs, batch).mean()
            loss_mse = mse_loss(outputs, batch)
            loss = 1.0 * loss_cos + 1.0 * loss_mse

            running_test_loss += loss.item() * batch.size(0)
            running_test_cosine_loss += loss_cos.item() * batch.size(0)
            running_test_mse_loss += loss_mse.item() * batch.size(0)

    epoch_test_loss = running_test_loss / len(test_loader.dataset)

    print(f"Epoch {epoch+1} | Train Loss: {epoch_train_loss:.6f} | Train Cosine Loss: {epoch_train_cosine_loss:.6f} | Train MSE Loss: {epoch_train_mse_loss:.6f} | "
          f"Test Loss: {epoch_test_loss:.6f} | Test Cosine Loss: {running_test_cosine_loss / len(test_loader.dataset):.6f} | Test MSE Loss: {running_test_mse_loss / len(test_loader.dataset):.6f}")

    torch.save(model, f"models/_/T5_AE_{epoch+1}.pt")
```
